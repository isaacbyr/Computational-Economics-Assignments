Consider the market for a homogeneous good (say, a flu vaccine). The demand function is q(p) = p^ε, where ε < 0 stands for the elasticity of demand. The supply function is 
q(p) = βp^η, where β > 0 is aconstant, and η > 0 stands for the elasticity of supply. Write Python codes to analyze this market in the following cases.

#### 2 a)
Set β = 1 and consider the following values for the elasticities: ε = −2 and η = 3. Plot the two indirect functions (note: the price has to be on the y-axis). Report the equilibrium price and quantity.

```python
# Creating an array of input values that I can pass into out functions for graphing
p = np.linspace(0.001,4, 200)
```

```python
# Next, I defined a function which takes the varibles e, B, n as inputs and then creates
# our supply and demand functions. Once the supply and demand functions are created
# the function returns them.

def create_function(e,B,n):
    qd = p**(e)  # Demand Function
    qs = B*p**n  # Supply Function
    
    return qd,qs

# This is the call to the create_function method to create our functions using e = -2,
# B = 1 and n = 3
qd, qs = create_function(-2, 1, 3)
```

```python
# I created another function which takes the supply and demand function as input
# and plots them on the same graph

def create_plot(qd, qs):

    fig, ax = plt.subplots(figsize = (10,8))
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    plt.xlim([0,15])

    plt.plot(qd,p, label = 'Demand')
    plt.plot(qs,p, label = 'Supply')
    plt.legend()

    plt.show()
  
    
    
# This is the call to create plot
create_plot(qd,qs)
```

```python
# To help me solve this I am going to use a library called sympy
from sympy import Symbol, Eq, solve

# start by defining all the varibles that we need
p = Symbol('p')
qd = Symbol('qd')
qs = Symbol('qd')

# Next we call the create_function again to give us our supply and demand functions
qd, qs = create_function(-2, 1, 3)

# Now that we have our supply and demand functions we put them into a sympy Eq and 
# store that equation as the varible eqn
eqn = Eq(qd, qs)

# We use a built sympy feature called solve to solve for p in our demand and supply functions
solve(eqn)

# This returns a list of number but we are only concered witht the first value as it
# is the optimal quantity
optimal_q = solve(eqn)[0]

# Next we can use the optimal quantity we calculated above and substite it into our demand
# function to get the optimal price
optimal_p = optimal_q**-2

print('The equilibrium price is: ', optimal_p)
print('The equilibrium quantity is: ',optimal_q)
```

#### b)
Now set β = 2 and consider the following values for the elasticities: ε = {−2; −0.5} and η = {1; 3}. Write a do-loop where at each iteration the equilibrium price and quantity are computed and displayed (for all four combinations). Report these four combinations in a table and comment

```python
df = pd.DataFrame(columns = ['Demand', 'Supply', "Eq Quantity", "Eq Price"])

p = Symbol('p')

for i in range(2): # This will index through the 2 values for demand elasticity
    for j in range(2): # This will index through the 2 values for supply elasticity
    
        # Set up a list for the demand elasticities that the first loop can iterate through
        e = [-2,-0.5]
        # Set up a list for the supply elasticities that the first loop can iterate through
        n = [1, 3]
        
        # This calls the create_function method again which creates the supply and demand 
        # functions using the corresponding elasticities. 
        qd, qs = create_function(e[i],2,n[j])

        # As I did above I will be using Sympy's solve method to help me solve these equations
        # Start by putting the supply and demand equations into the Sympy function Eq()
        eqn = Eq(qd, qs)
   
        # Now we can use Sympy's solve function to solve for the optimal q between supply and demand
        solve(eqn)
        
        # We are only concered with the first value as it is the optimal quantity
        optimal_q = solve(eqn)[0]
        
        # we can plug that optimal quantity into our demand function to get optimal price
        optimal_p = optimal_q ** e[i]
    
        # Load the demand, supply, and equilibrium price and quantity into a pandas dataframe
        df = df.append({'Supply': qs,"Eq Quantity": optimal_q, 'Demand': qd, 'Eq Price': optimal_p,
                       }, ignore_index=True)
                
# print pandas dataframe        
df
```

Above we get the results for the market equilibriums given the supply and demand functions.
